## 원뿔의 부피&겉넓이 계산 프로그램 5


```python
# 원뿔 부피 및 겉넓이 계산 프로그램

rad = int(input("반지름? : "))
hei = int(input("높이?: "))
sla = pow(rad ** 2 + hei ** 2, 0.5)
#부피 & 겉넓이 출력
print("주어진 원뿔의 부피는",1 / 3 * 3.14 * rad ** 2 * hei,"입니다" )
print("주어진 원뿔의 겉넓이는", 3.14 * rad ** 2 + 3.14 * rad * sla, "입니다")
```
